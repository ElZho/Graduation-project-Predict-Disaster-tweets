##Уберем url в отдельный столбец

```python
#create url column
def url_text(te_xt):
  url=re.compile(r'https?://\S+|www\.\S+')
  return url.match(te_xt)
data_union['url']=data_union['text'].apply(lambda x: str(url_text(x)))
```

##Вынесем hashtag в отдельный столбец

```python
#create hashtags column
def make_hash_col(te_xt):
  hashtag=str()
  l=len(re.findall(r'#', te_xt))
  for i in range(1, l+1):
    if i>1:
      hashtag=hashtag+',' +(te_xt.split(r"#")[i].split(' ')[0])
    else:
      hashtag=(te_xt.split(r"#")[i].split(' ')[0])
  return hashtag
data_union['hashtags']=data_union['text'].apply(lambda x:make_hash_col(x))
data_union['hashtags'].value_counts()
```

##Почистим столбец text от посторонних символов

Основная чистка - чистка основного столбца text

```python
def standardize_text(data):
    data['text']=data['text'].str.replace(r"http\S+", "")
    data['text']=data['text'].str.replace(r"http", "")
    data['text']=data['text'].str.replace(r"https", "")
    data['text']=data['text'].str.replace(r"@\S+", "")
    data['text']=data['text'].str.replace(r"[^A-Za-z0-9(),!?@\'\`\"\_\n]", " ")
    data['text']=data['text'].str.replace(r"@", "at")
    data['text']=data['text'].replace(r"#", " ")
    data['text']=data['text'].str.lower()
    data['text']=data['text'].replace(r'ûò', ' ')
    data['text']=data['text'].replace(r'\W', ' ')
    data['text']=data['text'].replace(r'_', " ")
    data['text']=data['text'].replace("'", "")
    return data

data_union = standardize_text(data_union)

# data_union.to_csv("clean_data.csv")
data_union.head()
```
